```python
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
for f in csv_files:
    df = pd.read_csv(f)
    plt.figure(figsize=(18,10))
    for i in range(len(df)):
        plt.plot(df.iloc[i][1:], label = df.iloc[i][0])
        plt.legend(bbox_to_anchor = (1.4, 1), loc='upper right')
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
for f in csv_files:
    df = pd.read_csv(f)
    plt.figure(figsize=(18,10))
    sns.set(rc={'axes.facecolor':'#eaeaf2','figure.facecolor':'#eaeaf2'})
    for i in range(len(df)):
        df_graph = sns.lineplot(df.iloc[i][1:], label = df.iloc[i][0])
        df_graph.set_xlabel("Months")
        df_graph.set_ylabel("AQI Index")
        plt.legend(bbox_to_anchor = (1.2, 1), loc='upper right')
```
